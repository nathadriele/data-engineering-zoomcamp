#### Total Rows in 2020 Yellow Taxi Data

Context: The Yellow Taxi database contains numerous entries throughout the year 2020.

Question: How many rows are there in the Yellow Taxi CSV files for the entire year of 2020?

#### SQL Query Used:

```python
SELECT 
  COUNT(*) 
FROM
  `ny-taxi-rides-447209.de_zoomcamp.yellow_tripdata`
WHERE 
  EXTRACT(YEAR FROM tpep_pickup_datetime) = 2020
  AND EXTRACT(YEAR FROM tpep_dropoff_datetime) = 2020;
```

```python
Answer: 24,648,499 rows
```

#### Rows in March 2021 Yellow Taxi Data

Context: Monthly-specific analyses provide more detailed insights into taxi data.

Question: How many rows are there in the Yellow Taxi CSV file for March 2021?

#### SQL Query Used:

```python
SELECT 
  COUNT(*) 
FROM
  `ny-taxi-rides-447209.de_zoomcamp.yellow_tripdata`
WHERE 
  EXTRACT(YEAR FROM tpep_pickup_datetime) = 2021
  AND EXTRACT(MONTH FROM tpep_pickup_datetime) = 03;
```

```python
Answer: 1,925,152 rows
```
